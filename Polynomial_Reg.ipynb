#Import Librari

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

#Import Dataset

```python
dataset = pd.read_csv('/content/Position_Salaries.csv')
X = dataset.iloc[:, 1].values
y = dataset.iloc[:, -1].values
```

```python
print(X)
```

```python
print(y)
```

#Traing Linear Reg Whole Datase

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X.reshape(-1, 1), y)
```

```python
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.scatter(X, y, color = 'red')
plt.plot(X, lin_reg.predict(X.reshape(-1, 1)), color = 'blue')
```

#Polynomial reg

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 4)
X_poly = poly_reg.fit_transform(X.reshape(-1, 1))
```

```python
poly_reg = LinearRegression()
poly_reg.fit(X_poly, y)
```

```python
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.scatter(X, y, color = 'red')

# Create a PolynomialFeatures object to transform X for plotting
# The degree should be the same as used for training (degree = 4)
poly_transformer_plot = PolynomialFeatures(degree = 4)
X_plot_poly = poly_transformer_plot.fit_transform(X.reshape(-1, 1))

# The previous error indicated that 'poly_reg' was a PolynomialFeatures object.
# This suggests the LinearRegression model was not correctly set or was overwritten.
# To ensure the plot works, we'll create and use a new LinearRegression model for prediction
# within this cell, assuming X_poly and y are available from previous steps.
from sklearn.linear_model import LinearRegression
local_poly_regressor = LinearRegression()
local_poly_regressor.fit(X_poly, y)

# Predict using the locally trained LinearRegression model
plt.plot(X, local_poly_regressor.predict(X_plot_poly), color = 'blue')
```
