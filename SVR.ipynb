#import lib

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

#import datset

```python
datset = pd.read_csv('Position_Salaries.csv')
X = datset.iloc[:, 1:-1].values
y = datset.iloc[:, -1].values
```

```python
print(X)
```

```python
print(y)
```

#feature scalining

```python
y = y.reshape(len(y), 1)
```

```python
print(y)
```

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
sc_y = StandardScaler()
```

```python
X = sc_X.fit_transform(X)
y = sc_y.fit_transform(y)
```

```python
print(X)
```

```python
print(y)
```

#train the model on the dataset

```python
from sklearn.svm import SVR
regressor = SVR()
regressor.fit(X, y)
```

#prediciting new result

```python
prediction = regressor.predict(sc_X.transform([[6.5]]))
print(prediction)
```

```python
prediction = sc_y.inverse_transform([[6.5]])
print(prediction)
```
